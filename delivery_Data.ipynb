```python
df = pd.read_csv('c3.csv')
df = pd.read_csv('c3.csv', parse_dates= ['time'])
data = pd.DataFrame(list(df['total deliveries']), index=df['time'], columns=['total deliveries'])
day = data.resample('d').sum()


mon = day['2022-10-17 0:00:00':'2022-11-05 0:00:00':7] # monday
tue = day['2022-10-18 0:00:00':'2022-11-05 0:00:00':7] # 
wed = day['2022-10-19 0:00:00':'2022-11-05 0:00:00':7] # 
thu = day['2022-10-20 0:00:00':'2022-11-05 0:00:00':7] # 
fri = day['2022-10-21 0:00:00':'2022-11-05 0:00:00':7] # 
sat = day['2022-10-22 0:00:00':'2022-11-05 0:00:00':7] # 
sun = day['2022-10-23 0:00:00':'2022-11-05 0:00:00':7] # sunday

A=np.mean(mon)
B=np.mean(tue)
C=np.mean(wed)
D=np.mean(thu)
E=np.mean(fri)
F=np.mean(sat)
G=np.mean(sun)


data = np.vstack((A, B, C, D, E, F, G))
day_averages = np.mean(data, axis=1)  # Calculate the average of deliveries for each day

days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

fig, ax = plt.subplots(figsize=(24, 8))
ax.bar(days, day_averages)

ax.set_xlabel('days')
ax.set_ylabel('deliveries')
ax.set_title('Bar chart showing the deliveries 2022-10-17 to 2022-11-05, Each bar represents average of 3 monday and simalarly tue, wed .. sun')

plt.show()
```

```python
df = pd.read_csv('c4.csv', parse_dates= ['time'])
df.set_index(['time'], inplace=True)
one_week = df['2022-10-15 00:00:00':'2022-11-16 23:00:00']

t=one_week['ti']
d=one_week['total deliveries']


data = {
    'time': t,
    'deliveries': d
}

df = pd.DataFrame(data)
df['time'] = pd.to_datetime(df['time'])
df['day'] = df['time'].dt.date
df['hour'] = df['time'].dt.hour

pivot_df = df.pivot_table(index='day', columns='hour', values='deliveries', aggfunc='sum')

plt.figure(figsize=(30, 16))
sns.heatmap(pivot_df, cmap='YlOrBr', annot=True, fmt='d')
plt.title('Heatmap of Deliveries showing each Day and each Hour')
plt.xlabel('Hour')
plt.ylabel('Day')
plt.show()          # sundays  16/10,  23/10,  30/10,  6/10,  13/10
```
